```python
# Import Library
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open the Data set
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
# Remove the categorical data(i.e str-state)
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
# splite the input & output columns
```

```python
dataset.columns
```

```python
independant=dataset[['Age', 'EstimatedSalary',  'Gender_Male']]
```

```python
independant
```

```python
independant.shape
```

```python
dependant= dataset[["Purchased"]]
```

```python
dependant
```

```python
#split the Train & Test dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independant,dependant,test_size=1/3,random_state=0)
```

```python
X_train
```

```python
# Train data-Model creation
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier =DecisionTreeClassifier(criterion='entropy',splitter='best',random_state= 0)
classifier.fit(X_train,Y_train)
```

```python
# Evaluation Method Presidcting the Test set result
```

```python

Y_pred= classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
```

```python
from sklearn.metrics import classification_report
clf_report =classification_report(Y_test,Y_pred)
```
